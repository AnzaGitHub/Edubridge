# Conclusion
### Students' performance were analysed in terms of different factors like gender, race, parent's qualification, 
### quality of the food they consumed and completion of test preparation courses prior to the exam.
### The effect of these factors on each tests: maths, reading and writing were analysed individually. Then the 
### mean of sum of scores for these tests were also considered. The effect of different factors in determining high 
### scores ie. above 60% were also analysed.
### All the analysis shows that, the factors like gender, race, parent's qlfcn, food quality and test preparation
### courses have a positive impact on improving the student's performance.
### In gender, female students scored a little better than their male counterparts. But in maths scores male 
### students outperformed female students. Also in the case of students below 60, males have better mean scores 
### than females.
### In race groups, group E always came at the top scorer position. Followed by group D, then group C, then 
### group B and at the lowest was group A. The only discrepancy happened is for students below 60, in which 
### case group D came at the top and group E descended down to 4th.
### In the case of parent's education level, students of parents with Master's degree always came at the top, 
### followed by Bachelor's degree, then Associate's degree, then some college, then some high school and at the
### lowest was high school.
### When the quality of food versus performance was analysed, students who consumed standard food always 
### faired better than students who consumed free/reduced food.
### Test preparation courses can also improve the performance. Students who had done a test preparation course
### always scored better than students who had not.
