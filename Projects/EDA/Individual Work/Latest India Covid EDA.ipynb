# Note 
#### When I tried to change the datatype I got error message.From the error I deduced that there is a string sign '-' present in the ConfirmedIndian and ConfirmedForeign columns. So firstly i need to replace '-' with 0.

### Checking for cells with '-' in the column ConfirmedIndian

```python
dwith_ = d[d["ConfirmedIndian"]=="-"]
dwith_
```

### Replacing '-' in the ConfirmedIndian and ConfirmedForeign column with 0. 

```python
d.replace({'ConfirmedIndian':{'-':0},'ConfirmedForeign':{'-':0}},inplace=True) # If more strings like ',','*' are present then 
#we can replace them too.For ex: d.replace({'ConfirmedIndian':{'-':0,',':0,'*':0},'ConfirmedForeign':{'-':0}},inplace=True) 
d
```

### Changing the data types of the ConfirmedIndian and ConfirmedForeign column 

```python
new_data_types={"ConfirmedIndian":int,"ConfirmedForeign":int}
d = d.astype(new_data_types)
d
```

### Statewise total of all cases 

```python
statewise=d.groupby(['State']).agg({"Confirmed":"max","Cured":"max","Deaths":"max"})                                                                                      
statewise
```

```python
statewise.index # checking the index of dataframe- statewise
```

### All case details of 5 states with most number of confirmed cases 

```python
most_confmd=statewise.sort_values(by='Confirmed',ascending=False).head(5)
most_confmd
```

### Pie plot for all cases of 5 states with most number of confirmed cases

```python
c=('crimson','forestgreen','lightblue','darkblue','purple')
most_confmd.plot(kind='pie', autopct='%1.1f%%', subplots = True,startangle=90,colors=c,shadow=True,explode=(.1,0,.2,0,0.3),
       legend=None,labels=('MH','KT','KL','TN','AP'))
fig = plt.gcf()
fig.set_size_inches(20,7)
plt.show()
```

### Total confirmed cases in India

```python
confmd=statewise.Confirmed.sum()
confmd
```

### Total cured cases in India 

```python
cured=statewise.Cured.sum()
cured
```

### Cure Rate 

```python
CureRate=cured/confmd*100
CureRate
```

### Total deaths in India 

```python
deaths=statewise.Deaths.sum()
deaths
```

### Death Rate 

```python
DeathRate=deaths/confmd*100
DeathRate
```

### Finding rates of each cases against the national total 

```python
ConfirmedVsNational=[]
CuredVsNational=[]
DeathsVsNational=[]
for i in range(len(statewise)):
    ConNat=statewise.Confirmed[i]/confmd*100
    ConfirmedVsNational.append(ConNat)
    CuNat=statewise.Cured[i]/cured*100
    CuredVsNational.append(CuNat)
    DNat=statewise.Deaths[i]/deaths*100
    DeathsVsNational.append(DNat)

x=statewise.iloc[:,:0]
NatRate=x

NatRate['Confirmed Vs National']=ConfirmedVsNational
NatRate['Cured Vs National']=CuredVsNational
NatRate['Deaths Vs National']=DeathsVsNational
NatRate
```

### 5 states with most number of confirmed cases 

```python
most_confirmed=NatRate.sort_values(by='Confirmed Vs National',ascending=False).head(5)
most_confirmed
```

### Bar plot for all cases of 5 states with most number of confirmed cases 

```python
most_confirmed.plot.bar(color=['crimson','forestgreen','grey'],figsize=(15,5),width=1)
plt.xlabel("States",size=15)
plt.ylabel('Cases',size=15)
plt.show()
```

# Summary
#### From the above grouping and visualizations, we can see that Maharastra has the most number of confirmed cases, followed by Kerala, Karnataka, Tamil Nadu and Andhra Pradesh. So COVID 19 has affected the central and south regions of India more when  compared  to the north.
